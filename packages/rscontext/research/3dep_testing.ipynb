```python
import json
from shapely.geometry import shape, GeometryCollection, Polygon
import py3dep
import matplotlib.pyplot as plt
```

```python
with open(r"test_boundary.geojson") as f:
    features = json.load(f)['features']
geom_boundary = GeometryCollection([shape(features[0]["geometry"])])[0]
geom_boundary
```

Resolution of 10m for this catchment is successful, however this previouly resulted in an error (same message as 1m resolution, see below), and currently occurs on the huc8 watershed of this catchment.

```python
dem_10m = py3dep.get_map("DEM", geom_boundary, resolution=10, geo_crs="epsg:4326")
plt.imshow(dem_10m.values,cmap='terrain')
plt.show()
```

Resolution of 1.1m is flipped vertically. This has happend on occasion with other catchments, but I cannot find a pattern.

```python
dem_1_1m = py3dep.get_map("DEM", geom_boundary, resolution=1.1, geo_crs="epsg:4326")
plt.imshow(dem_1_1m.values,cmap='terrain')
plt.show()
```
